```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

vehicle_age = np.random.uniform(0,10,1000)
vehicle_price = 70000 - (vehicle_age + np.random.normal(0, 1, 1000)) * 3000

vehicle_age = vehicle_age.reshape(-1, 1)  #reshape for single feature, use (1,-1) for single sample
vehicle_price = vehicle_price.reshape(-1, 1)

plt.scatter(vehicle_age, vehicle_price)
reg = LinearRegression().fit(vehicle_age, vehicle_price)
reg.score(vehicle_age, vehicle_price)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

vehicle_age = np.random.uniform(0,10,1000)
vehicle_price = 70000 - (vehicle_age + np.random.normal(0,10,1000)) * 3000

vehicle_age = vehicle_age.reshape(-1, 1)  #reshape for single feature, use (1,-1) for single sample
vehicle_price = vehicle_price.reshape(-1, 1)

plt.scatter(vehicle_age, vehicle_price)
reg = LinearRegression().fit(vehicle_age, vehicle_price)
reg.score(vehicle_age, vehicle_price)
```

1 -> Standard Deviation = 1: R-squared = 0.8964
The R-squared value of 0.8964 shows that 89.64% of the variability in vehicle price is explained by vehicle age. The standard deviation of 1 added to the price using np.random.normal(0, 1, 1000) is small, so the vehicle price remains closely aligned with vehicle age. As a result, the regression model fits the data well, indicating that vehicle age is a strong predictor of vehicle price in this case.

2 -> Standard Deviation = 5: R-squared = 0.2864
The R-squared value of 0.2864 indicates that only 28.64% of the variability in vehicle price is explained by vehicle age in this model. This lower value, compared to the previous example, is due to the increased standard deviation of 5 added to the vehicle price using np.random.normal(0, 5, 1000). As a result, the model’s ability to explain the price variability significantly decreases, leading to a lower R-squared value.

3 -> Standard Deviation = 10: R-squared = 0.0810
The R-squared value of 0.0810 indicates that only 08.10% of the variability in vehicle price is explained by vehicle age. The significant decrease in the R-squared value compared to previous cases is due to the larger standard deviation of 10 added to the vehicle price using np.random.normal(0,10,1000).As a result, the model performs poorly in explaining the variation, and most of the price variability remains unexplained.

As the standard deviation increases, the R-squared value decreases. This happens because higher randomness added to the data, thus, obscuring the clear linear relationship between vehicle age and price.
